```python
%matplotlib inline

from sklearn import datasets, svm
import matplotlib.pyplot as plt
import pickle #use joblib instead of pickle. It's more efficient!
from sklearn.externals import joblib
iris = datasets.load_iris()
digits = datasets.load_digits()
```

```python
dir2 = lambda obj: [x for x in dir(obj) if x[:2] != '__']
keys = lambda dictionary: [key for key in dictionary]
```

```python
print("Iris features: ", keys(iris))
print("Digits features: ", keys(digits))
```

```python
clf = svm.SVC(gamma=0.001, C=100.)
```

```python
clf.fit(digits.data[:-1], digits.target[:-1]) 
```

```python
clf.predict(digits.data[-1:])
```

```python
plt.matshow(digits.images[-1], cmap=plt.cm.Greys)
```

```python
X, y = iris.data, iris.target
clf.fit(X, y)
```

```python
s = pickle.dumps(clf)
clf2 = pickle.loads(s)
clf2.predict(X[0:1])
```
